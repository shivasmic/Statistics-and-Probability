# Data Visualization Techniques

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1,4,9,16],color = 'red', linewidth = 4.0)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
line, = plt.plot([1,2,3,4],[1,4,9,16])
line.set_linestyle('--')
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.setp(line,color='r',linewidth=2.0)
line.set_linestyle('--')
plt.show()
```

```python
#Multi-plot
#with plt.style.context(('dark_background')):
p1 = np.arange(0.0,30.0,0.1)
plt.subplot(211)#Two rows 1 column and two plots
plt.plot(p1,np.sin(p1)/p1,'b--')
plt.subplot(212)
plt.plot(p1,np.cos(p1)/p1,'r--')
plt.show()
```

```python
plt.style.use('ggplot')
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
with plt.style.context(('dark_background')):
    plt.plot([1,2,3,4],[1,4,9,16])
    plt.show()
```
